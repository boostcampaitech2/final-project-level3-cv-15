```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

cv2.contourArea(contour, oriented=None) -> retval

- contour: 외곽선 좌표. numpy.ndarray. shape=(K, 1, 2)
- oriented: True이면 외곽선 진행 방향에 따라 부호 있는 면적을 반환. 기본값은 False.
- retval: 외곽선으로 구성된 영역의 면적

```python
mycoord = []
mycoord.append([0,0])
mycoord.append([10,0])
mycoord.append([10,10])
# mycoord.append([0,10])
temp_array = np.zeros([50, 50])
# mycoord.append([845, 0])
# mycoord.append([845, 1080])
# mycoord.append([1920, 1080])
# mycoord.append([1920, 0])

# temp_array = np.zeros([1081, 1921])

line_image = cv2.fillPoly(temp_array, [np.array(mycoord)],255)

# line_image = cv2.polylines(temp_array, [np.array(mycoord)], False, 255, 1) # contourArea 에 인자로 넣어줄때는 False, 그릴때만 True
cv2.contourArea(np.argwhere(line_image==255))
print(f'# pixel : {len(np.argwhere(line_image==255))}')
print(f'dim : {np.argwhere(line_image==255).shape}')
plt.imshow(line_image)
print(f'area : {cv2.contourArea(np.argwhere(line_image==255))}')
print(f'proportion : {len(np.argwhere(line_image==255))/(1920*1080)*100}')
# print(mycoord)
```
